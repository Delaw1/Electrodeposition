```python
@interact
def heating_power(acdc_voltage = 500, # V
                  acdc_current = 3.5, # A
                  acdc_number = 3,
                  acdc_efficiency = 0.97,
                  dcdc_voltage = 48., # V
                  dcdc_current = 21., # A
                  dcdc_number = 6,
                  dcdc_efficiency = 0.85,
                 ):
    """
    The rate of generation of heat that needs to be dissipated, in W
    """
    acdc_power = acdc_voltage * acdc_current * acdc_number
    heating_power = acdc_power * (1.-acdc_efficiency)
    dcdc_power = dcdc_voltage * dcdc_current * dcdc_number
    heating_power += dcdc_power * (1.-dcdc_efficiency)
    return heating_power

```

```python
@interact
def hull_surface_area(hull_diameter = 0.5, #m
                      hull_length = 3.0 # m
                     ):
    """
    The external surface area of the hull. m^2
    """
    return np.pi * (hull_diameter * hull_length + 
                      2 * hull_diameter*hull_diameter / 4.)
    
```

## External heat transfer
External flow over the cylinder is not quite like a sphere, but this will at least give us a guess to see whether it matters. From [Perry's Chemical Engineers' Handbook, Eigth Edition](http://www.accessengineeringlibrary.com/browse/perrys-chemical-engineers-handbook-eighth-edition), chapter 5.3:
For an isothermal spherical surface, Whitaker recommends [AIChE, 18, 361 (1972)]

$$\bar{Nu_D} = 2 + (0.4 Re_D^{1/2} + 0.06 Re_D^{2/3})Pr^{0.4} \left(\frac{\mu_e}{\mu_s}\right) ^{1/4}$$
(5-57)

This equation is based on data for 0.7 < Pr < 380, 3.5 < ReD < 8 × 104, and 1 < (μe/μs) < 3.2. The properties are evaluated at the free-stream temperature Te, with the exception of μs, which is evaluated at the surface temperature Ts.

$\bar{Nu} = \bar{h}D/k$ for 
$q = h (T_{surface} - T_{fluid})$ where h [W/(m2·K)] is the local heat-transfer coefficient and q is the energy flux (W/m2)

### Seawater properties
Temperature
8-16 C from https://www.cefas.co.uk/cefas-data-hub/sea-temperature-and-salinity-trends/presentation-of-results/station-32-angle/

Salinity 34.5

```python
@interact(temperature=10)
def prandtl_seawater(temperature):
    """
    Gives the Prandtl number of seawater at the specified temperature in celsius.
    
    Prandtl number Pr = ν/α (kinematic viscosity) over (thermal diffusivity)
    
    from http://web.mit.edu/seawater/Seawater_Property_Tables_8March16.pdf
    Assuming a salinity of 30 because the following suggests it's about 34.5
    https://www.cefas.co.uk/cefas-data-hub/sea-temperature-and-salinity-trends/
    
    May be wise to interpolate on salinity too, if this matters.
    """
    T = [0, 10, 20, 30, 40]
    Pr = [13.31, 9.48, 7.12, 5.57, 4.49]
    interpolation = scipy.interpolate.interp1d(T, Pr, kind="quadratic")
    return float(interpolation(temperature))

```

```python
@interact(temperature=10)
def density_seawater(temperature):
    """
    Return the density of seawater, kg/m3, at specified T (celsius).
    
    from http://web.mit.edu/seawater/Seawater_Property_Tables_8March16.pdf
    Assuming a salinity of 30 because the following suggests it's about 34.5
    https://www.cefas.co.uk/cefas-data-hub/sea-temperature-and-salinity-trends/
    
    May be wise to interpolate on salinity too, if this matters.
    """
    T = [0, 10, 20, 30, 40]
    rho = [1024.0, 1023.0, 1021.1, 1018.2, 1014.6]
    interpolation = scipy.interpolate.interp1d(T, rho, kind="quadratic")
    return float(interpolation(temperature))

```

```python
@interact(temperature=10)
def thermal_conductivity_seawater(temperature):
    "Similar to above, but for k in (W/m K)"
    T = [0, 10, 20, 30, 40]
    k = [ 0.570, 0.587, 0.602, 0.616, 0.628]
    interpolation = scipy.interpolate.interp1d(T, k, kind="quadratic")
    return float(interpolation(temperature))

```

```python
@interact(temperature=10)
def dynamic_viscosity_seawater(temperature):
    "Similar to above but for mu in (kg/m s)"
    T = [0, 10, 20, 30, 40]
    mu = 1e-3* np.array([1.887, 1.382, 1.065, 0.851, 0.699])
    interpolation = scipy.interpolate.interp1d(T, mu, kind="quadratic")
    return float(interpolation(temperature))

dynamic_viscosity_seawater(15)
```

```python
@interact(velocity = 12, lengthscale=0.5, temperature=10)
def reynolds_seawater(velocity, lengthscale, temperature):
    """
    Reynolds number, for thing of lengthscale at velocity in 
    seawater at specified temperature in celsius.
    
    Re = G D / mu # where G is the mass velocity
    Re = rho v D / mu
    
    ρ is the density of the fluid (SI units: kg/m3)
    v is a characteristic velocity of the fluid with respect to the object (m/s)
    L is a characteristic linear dimension (m)
    μ is the dynamic viscosity of the fluid (Pa·s or N·s/m2 or kg/(m·s)).
    """
    return density_seawater(temperature) * velocity * lengthscale / dynamic_viscosity_seawater(temperature)
```

Sanity check on Reynolds number:
> The Reynolds Number Re based on the total length of Collins traveling at 10 knots in standard sea water is 3.2 × 10$^8$

from "Some Aspects of Submarine Design" by Prof. P. N. Joubert, report DSTO–TR–1920 for the Australian Government Department of Defense


```python
Re = reynolds_seawater(velocity=5.14, lengthscale=77, temperature=13)
print("By our estimate, Re = {:.2g} for a 77m Collins submarine at 10 knots in 13˚C water".format(Re))
```

Returning to the heat transfer calculation:


Nusselt number:
$$\bar{Nu_D} = 2 + (0.4 Re_D^{1/2} + 0.06 Re_D^{2/3})Pr^{0.4} \left(\frac{\mu_e}{\mu_s}\right) ^{1/4}$$
(5-57)

```python
@interact(velocity=12,diameter=0.5, water_temperature=10)
def nusselt(velocity, diameter, water_temperature):
    """
    Nusselt number for a sphere of given diamater, at given velocity, in seawater.
    
    water_temperature in Celsius
    """
    
    temperature_drop = 5 # guess, to avoid implicit or recursive solve
    surface_temperature = water_temperature + temperature_drop
    
    Re = reynolds_seawater(velocity, lengthscale=diameter, temperature=water_temperature)
    Pr = prandtl_seawater(temperature=water_temperature)
    viscosity_factor = ( dynamic_viscosity_seawater(water_temperature) / 
                         dynamic_viscosity_seawater(surface_temperature)) ** 0.25
    
    assert viscosity_factor<1.05, "Viscocity change altering Nu by more than 5%, please check the temperature drop assumption!"
    
    if not 0.7 < Pr < 380:
        print("Warning, extrapolating to Pr={:.2g} where formula not tested".format(Pr))
    if not 3.5 < Re < 8e4:
        print("Warning, extrapolating to Re={:.2g} where formula not tested".format(Re))
        if Re>8e4:
            Re=8e4
            print("Instead capping Re=8e4, which will underestimate the Nu number",
                  "(overestimate the temperature drop outside the hull, and thus",
                  "overestimate the temperature inside the hull)")
    if not 1 < viscosity_factor**4 < 3.2:
        print("Warning, extrapolating viscosity ratio where formula not tested")
    
    Nu = 2 + (0.4 * Re**0.5 + 0.06 * Re**(2./3.)) * viscosity_factor
    return Nu
```

```python
@interact(velocity=12,diameter=0.5, water_temperature=10)
def heat_transfer_coefficient(velocity, diameter, water_temperature):
    """
    h = Nu k / D
    
    Nu is dimensionless
    k is heat transfer coefficient in W/m/K
    D is diameter in m
    h is in W/m2/K
    """
    Nu = nusselt(velocity, diameter, water_temperature)
    k = thermal_conductivity_seawater(water_temperature)
    h = Nu * k / diameter
    return h
```

```python
@interact(hull_diameter=0.5, hull_length=3, velocity=12, water_temperature=10, heating_power=HEATING_POWER)
def external_hull_surface_temperature(hull_diameter,
                        hull_length,
                        velocity,
                        water_temperature,
                        heating_power,
                       ):
    
    """
    The temperature on the external surface of the hull.
    """


    """
    We calculated the h assuming it was a sphere, 
    so perhaps should continue to model it that way,
    but for now lets switch to it being a cylinder.
    If length makes a significant difference, revisit this sphere assumption.
    """
    hull_surface_area_ = hull_surface_area(hull_diameter, hull_length)
    
    heat_flux = heating_power / hull_surface_area_
    
    heat_transfer_coefficient_ = heat_transfer_coefficient(velocity,
                                                          hull_diameter,
                                                          water_temperature)
    
    temperature_drop = heat_flux / heat_transfer_coefficient_
    
    surface_temperature = water_temperature + temperature_drop
    return surface_temperature    
```

So far it looks like the external heat trasfer is fast, the temperature drop across the thermal boundary layer in the sea water is less than 1 degree, and so calculating it to even this level of detail was more than sufficient.


## Heat transfer through the wall

Conduction through the steel hull wall, to find the temperature on the inside surface

```python
"""
Some thermal conductivities in W/(m K) from 
http://www.engineeringtoolbox.com/thermal-conductivity-d_429.html
"""
thermal_conductivities = {
    'steel (carbon)': 43.,
    'steel (stainless)': 16.,
    'aluminium': 205.,
}
```

```python
@interact(hull_diameter=0.5, hull_length=3, velocity=12, water_temperature=10,
         thermal_conductivity_hull=43, hull_thickness=(0,0.1,0.01), heating_power=HEATING_POWER)

def internal_hull_surface_temperature(
                                     hull_diameter,
                                     hull_length,
                                     velocity,
                                     water_temperature, # C
                                     thermal_conductivity_hull, # W/m/K 
                                     hull_thickness, # m
                                     heating_power, # W
                                     ):
    """
    The temperature in celsius on the inside of the hull.
    """
    
        
    external_temperature_ = external_hull_surface_temperature(
                        hull_diameter,
                        hull_length,
                        velocity,
                        water_temperature,
                        heating_power
                       )

    hull_surface_area_ = hull_surface_area(hull_diameter, hull_length) # m2
    
    heat_flux = heating_power / hull_surface_area_
    

    wall_temperature_drop = heat_flux * hull_thickness / thermal_conductivity_hull
    
    internal_temperature = external_temperature_ + wall_temperature_drop
    
    return internal_temperature

    
```

It's still looking pretty cool on the inside of the hull wall.
But now we reach the gas, which is much less thermally conductive...

## Heat transfer inside the hull

Now we've solved for the temperature of the inside wall of the hull, we need to figure out heat transfer from the heat-generating hot electrical components, to the wall. This will occur via simultaneous convection through the (nitrogen) gas, and conduction along the (aluminium) beam.

### Heat transfer through the gas by natural convection.

From Perry's chapter 5, again:
 In addition to the Nusselt number mentioned above, the key dimensionless parameters for natural convection include the Rayleigh number $Ra_x = β ΔT g x^3 / ν α$ and the Prandtl number $Pr = ν/α$. The properties appearing in Ra and Pr include the volumetric coefficient of expansion β (K$^{−1}$); the difference ΔT between the surface (Ts) and free stream (Te) temperatures (K or °C); the acceleration of gravity g (m/s2); a characteristic dimension $x$ of the surface (m); the kinematic viscosity ν (m2/s); and the thermal diffusivity α (m2/s). The volumetric coefficient of expansion for an ideal gas is β = 1/T, where T is absolute temperature.
 
 Enclosed Spaces The rate of heat transfer across an enclosed space.
 
 For a horizontal rectangular cavity heated from below, the onset of advection requires $Ra_L > 1708$. Globe and Dropkin [J. Heat Transfer, 81, 24–28 (1959)] propose the correlation
$$\bar{Nu_L} = 0.069 Ra_L^{1/3} Pr^{0.074}$$  for 3e5 < $Ra_L$ < 7e9.
(5-42)

All properties in (5-42) are calculated at the average temperature $(T_H + T_C)/2$.

    

```python
def kinematic_viscosity_nitrogen(temperature_C = 10, pressure_bar = 10):
    """
    get the kinematic viscosity at the temperature in C 
    and pressure in bar
    returns result in m2/s
    """
    fluid = 'Nitrogen'
    T = 273.15 + temperature_C
    P = 1e5 * pressure_bar
    dynamic_viscosity = CoolProp.PropsSI('viscosity','T',T, 'P',P,fluid)
    #print(dynamic_viscosity)
    density = CoolProp.PropsSI('Dmass','T',T, 'P',P,fluid)
    #print(density)
    kinematic_viscosity = dynamic_viscosity / density
    #print(kinematic_viscosity)
    return kinematic_viscosity
print("{:.4g} should be about 1.4474e-6 m2/s".format(kinematic_viscosity_nitrogen(10., 10.))) # according to http://www.peacesoftware.de/einigewerte/stickstoff_e.html
print("{:.4g} should be about 1.6194e-7 m2/s".format(kinematic_viscosity_nitrogen(10., 100.))) # according to http://www.peacesoftware.de/einigewerte/stickstoff_e.html
```

```python
def thermal_diffusivity_nitrogen(temperature_C = 10, pressure_bar = 10):
    """
    get the thermal diffusivity at the temperature in C 
    and pressure in bar
    returns result in m2/s
    """
    fluid = 'Nitrogen'
    T = 273.15 + temperature_C
    P = 1e5 * pressure_bar
    thermal_conductivity = CoolProp.PropsSI('conductivity','T',T, 'P',P,fluid) # W/m/K
    density = CoolProp.PropsSI('Dmass','T',T, 'P',P,fluid)
    specific_heat_capacity = CoolProp.PropsSI('Cpmass','T',T, 'P',P,fluid)
    thermal_diffusivity = thermal_conductivity / (density * specific_heat_capacity)
    return thermal_diffusivity
print("{:.4g} should be about 2.0126e-6 m2/s".format(thermal_diffusivity_nitrogen(10., 10.))) # according to http://www.peacesoftware.de/einigewerte/stickstoff_e.html
print("{:.4g} should be about 2.094e-7 m2/s".format(thermal_diffusivity_nitrogen(10., 100.))) # according to http://www.peacesoftware.de/einigewerte/stickstoff_e.html
```

```python
@interact
def internal_nusselt(temperature_hot = 50, # C
                     temperature_cold = 10, # C 
                     effective_gravity = 10, # m/s2
                     spacing = 0.1, # m
                     pressure_bar = 10, # bar
                    ):
    """
    
    temperature_hot is the hot surface, in degrees celsius
    temperature_cold is the cold, i.e. the inside of the hull (also in celsius)
    effective_gravity = centripetal force? (would be 9.81 m/s2 if stationary)
    spacing [m] width of cavity. 
        Note: For rectangular cavities, the plate spacing 
        between the two surfaces L is the characteristic dimension that 
        defines the Nusselt and Rayleigh numbers. 
    
    Nu = h L / k
    """
    average_temperature = (temperature_hot + temperature_cold) / 2.
    "Temperatures should be in celsius"
    average_temperature_kelvin = average_temperature + 273.15
    
    delta_T = temperature_hot - temperature_cold
    assert delta_T > 0, "Hot should be hotter than cold"
        
    # beta
    volumetric_coefficient_of_expansion  = 1. / average_temperature_kelvin # for ideal gas
    
    kinematic_viscosity = kinematic_viscosity_nitrogen(temperature_C=average_temperature, pressure_bar=pressure_bar)
    thermal_diffusivity = thermal_diffusivity_nitrogen(temperature_C=average_temperature, pressure_bar=pressure_bar)

    # Pr = ν/α (kinematic viscosity) over (thermal diffusivity)
    prandtl = kinematic_viscosity / thermal_diffusivity
        
    rayleigh = float(volumetric_coefficient_of_expansion * delta_T * effective_gravity * spacing**3 / (kinematic_viscosity * thermal_diffusivity))
    if rayleigh < 1708: 
        print("Warning: Rayleigh number too low for onset of advection. No convection")
        return 0.
    if not 3e5 <  rayleigh < 7e9:
        print(rayleigh)
        print("Warning, correlation only tested for Rayleigh number between 3e5 and 7e9, not {:.0g}".format(rayleigh))

    
    nusselt = 0.069 * rayleigh**(1./3) * prandtl**0.074
    return nusselt
internal_nusselt()
```

```python
@interact(spacing=(0.,1.,0.02))
def internal_gas_heat_transfer_coefficient(
                     temperature_hot = 50, # C
                     temperature_cold = 10, # C 
                     effective_gravity = 10, # m/s2
                     spacing = 0.1, # m
                     pressure_bar = 10., # bar
                    ):
    """
    Returns the heat transfer coefficient h in W/m2/K
    
    For thermal flux q = h * (Thot - Tcold)
    
    From Nu = h L / k, and taking Nu from the internal_nusselt() function.
    """
    
    temperature_C = (temperature_hot + temperature_cold)/2.
    fluid = 'Nitrogen'
    T = 273.15 + temperature_C
    P = 1e5 * pressure_bar
    thermal_conductivity = CoolProp.PropsSI('conductivity','T',T,'P',P,fluid) # W/m/K
    
    nusselt = internal_nusselt(temperature_hot = temperature_hot, # C
                               temperature_cold = temperature_cold, # C 
                               effective_gravity = effective_gravity, # m/s2
                               spacing = spacing, # m
                               pressure_bar = pressure_bar, # bar
                               )
    
    h = nusselt * thermal_conductivity / spacing
    return h
internal_gas_heat_transfer_coefficient()
```

```python
@interact
def thermal_resistance_internal_gas(temperature_hot = 20, # C
                                    temperature_cold = 10, # C
                                    effective_gravity = 10, # m/s2
                                    shelf_spacing = 0.1, # m
                                    shelf_length = 1.0, # m
                                    shelf_depth = 0.1, # m
                                    pressure_bar = 10, # bar
                                    ):
    """
    The resistance to heat transfer across the gas gap 
    inside the device. Returns value in K/W
    """
    
    heat_transfer_coefficient = internal_gas_heat_transfer_coefficient(
                                    temperature_hot = temperature_hot, # C
                                    temperature_cold = temperature_cold, # C
                                    effective_gravity = effective_gravity, # m/s2
                                    spacing = shelf_spacing, # m
                                    pressure_bar = pressure_bar, # bar
    )
    surface_area = shelf_length * shelf_depth
    thermal_resistance = 1. / (surface_area * heat_transfer_coefficient)
    return thermal_resistance
    
#thermal_resistance_internal_gas()
```

```python

@interact(hull_diameter=0.5, hull_length=3, velocity=12, water_temperature=10,
         thermal_conductivity_hull=43, hull_thickness=(0,0.1,0.01),
          effective_gravity=(0,100.), shelf_spacing=(0,1.,0.02),
          shelf_length=(0,2.,0.02), shelf_depth=(0,1.,0.02),
          pressure_bar=10, heating_power=(0,2*HEATING_POWER)
         )

def component_temperature_gas_only(
                                     hull_diameter,
                                     hull_length,
                                     velocity,
                                     water_temperature,
                                    heating_power,
                                     thermal_conductivity_hull, # W/m/K 
                                     hull_thickness, # m
                                    effective_gravity = 10, # m/s2
                                    shelf_spacing = 0.1, # m
                                    shelf_length = 1.0, # m
                                    shelf_depth = 0.2, # m
                                    pressure_bar = 10, # bar
                                     ):

    internal_hull_surface_temperature_ = internal_hull_surface_temperature(
                                     hull_diameter,
                                     hull_length,
                                     velocity,
                                     water_temperature,
                                     thermal_conductivity_hull, # W/m/K 
                                     hull_thickness, # m
                                     heating_power, # W
                                     )

        
    def error_in_hot_temperature(temperature_hot):
        if temperature_hot < internal_hull_surface_temperature_:
            return temperature_hot - internal_hull_surface_temperature_ + error_in_hot_temperature(internal_hull_surface_temperature_+0.1)

        thermal_resistance = thermal_resistance_internal_gas(
                                    temperature_hot = temperature_hot, # C
                                    temperature_cold = internal_hull_surface_temperature_, # C
                                    effective_gravity = effective_gravity, # m/s2
                                    shelf_spacing = shelf_spacing, # m
                                    shelf_length = shelf_length, # m
                                    shelf_depth = shelf_depth, # m
                                    pressure_bar = pressure_bar, # bar
                                    ) # [K/W]
        delta_T = heating_power * thermal_resistance
        #print(temperature_hot, delta_T)
        error = internal_hull_surface_temperature_ + delta_T - temperature_hot
        return error
    
    # initial guess
    temperature_hot = 80 # C
    
    solution = scipy.optimize.root(error_in_hot_temperature, temperature_hot)
    if solution.success:
        temperature_hot = solution.x[0]
    else:
        raise RuntimeError(solution.message)
    
    #T = range(11,249,10)
    #e = [error_in_hot_temperature(t) for t in T]
    #plt.plot(T,e)
    
    return temperature_hot


```

It's looking like with only gas convection, the hot components reach about 200C - much too hot. And quite sensitive to a bunch of guesses.

## Conduction along the beam



```python
@interact(beam_thickness=(0,0.2,0.01))
def thermal_resistance_along_beam(
        thermal_conductivity_beam = 205, # W/m/K, aluminium
        beam_thickness = 0.02, # m, perpendicular to heat flux
        beam_length = 0.5, # m,  the distance along which the heat must diffuse
        beam_height = 0.5, # m, perpendicular to heat flux
        ):
    """
    The resistance to heat transfer along the aluminium beam 
    inside the device. Returns value in K/W
    """
    cross_sectional_area = beam_height * beam_thickness
    thermal_resistance = beam_length / (thermal_conductivity_beam * cross_sectional_area)
    return thermal_resistance
```

## Conduction and convection

```python
@interact(hull_diameter=0.5, hull_length=3, velocity=12, water_temperature=10,
         thermal_conductivity_hull=43, hull_thickness=(0,0.1,0.01),
          effective_gravity=(0,100.), shelf_spacing=(0,1.,0.02),
          shelf_lengthself_width=(0,2.,0.02), shelf_depth=(0,1.,0.02),
          pressure_bar=10, beam_thickness=(0,0.2,0.01)
         )

def component_temperature_gas_and_beam(
                                     hull_diameter,
                                     hull_length,
                                     velocity,
                                     water_temperature,
                                     thermal_conductivity_hull, # W/m/K 
                                     hull_thickness, # m
                                    effective_gravity = 10, # m/s2
                                    shelf_spacing = 0.1, # m
                                    shelf_length = 1.0, # m
                                    shelf_depth = 0.2, # m
                                    pressure_bar = 10, # bar
                                    thermal_conductivity_beam=205, # W/m/K, aluminium
                                    beam_thickness = 0.02, # m, perpendicular to heat flux
                                    beam_length = 0.5, # m,  the distance along which the heat must diffuse
                                    beam_height = 0.5, # m
                                      acdc_voltage = 500, # V
                                      acdc_current = 3.5, # A
                                      acdc_number = 3,
                                      acdc_efficiency = 0.97,
                                      dcdc_voltage = 48., # V
                                      dcdc_current = 21., # A
                                      dcdc_number = 6,
                                      dcdc_efficiency = 0.85,
                                     ):

    heating_power_ = heating_power(acdc_voltage = acdc_voltage, # V
              acdc_current = acdc_current, # A
              acdc_number = acdc_number,
              acdc_efficiency = acdc_efficiency,
              dcdc_voltage = dcdc_voltage, # V
              dcdc_current = dcdc_current, # A
              dcdc_number = dcdc_number,
              dcdc_efficiency = dcdc_efficiency,
             ) # [W]

    internal_hull_surface_temperature_ = internal_hull_surface_temperature(
                                     hull_diameter = hull_diameter,
                                     hull_length = hull_length,
                                     velocity = velocity,
                                     water_temperature = water_temperature,
                                     thermal_conductivity_hull = thermal_conductivity_hull, # W/m/K 
                                     hull_thickness = hull_thickness, # m
                                     heating_power = heating_power_,
                                     )
    

        
    def error_in_hot_temperature(temperature_hot):
        if temperature_hot < internal_hull_surface_temperature_:
            return temperature_hot - internal_hull_surface_temperature_ + error_in_hot_temperature(internal_hull_surface_temperature_)

        thermal_resistance_gas_ = thermal_resistance_internal_gas(
                                    temperature_hot = temperature_hot, # C
                                    temperature_cold = internal_hull_surface_temperature_, # C
                                    effective_gravity = effective_gravity, # m/s2
                                    shelf_spacing = shelf_spacing, # m
                                    shelf_length = shelf_length, # m
                                    shelf_depth = shelf_depth, # m
                                    pressure_bar = pressure_bar, # bar
                                    ) # [K/W]
        
        thermal_resistance_beam_ = thermal_resistance_along_beam(
                                    thermal_conductivity_beam=thermal_conductivity_beam,
                                    beam_thickness=beam_thickness,
                                    beam_length=beam_length,
                                    beam_height=beam_height)
        
        # the convection and beam transfer are in parallel
        thermal_resistance = 1./(1./thermal_resistance_gas_ + 1./thermal_resistance_beam_)
        delta_T = heating_power_ * thermal_resistance
        #print(temperature_hot, delta_T)
        error = internal_hull_surface_temperature_ + delta_T - temperature_hot
        return error
    
    # initial guess
    temperature_hot = 50 # C
    
    solution = scipy.optimize.root(error_in_hot_temperature, temperature_hot)
    if solution.success:
        temperature_hot = solution.x[0]
    else:
        raise RuntimeError(solution.message)
    
    #T = range(11,249,10)
    #e = [error_in_hot_temperature(t) for t in T]
    #plt.plot(T,e)
    
    return temperature_hot
```

We have a temperature, with convection and conduction!

And it's above 50˚C!

```python
def run_many_simulations(data):
    results = []
    with open('simulations.log', 'w') as f:
        for i,row in enumerate(data):
            with redirect_stdout(f):
                values_dict = {p.name: v for p, v in zip(variables, row)}
                values_dict.update(fixed_parameter_values)
                print(values_dict)
                try:
                    result = component_temperature_gas_and_beam(**values_dict)
                except RuntimeError:
                    result = np.nan
                results.append(result)
                print(result)
                print('-'*80)
            print("Run {} of {} gives {}".format(i+1, len(data), result))
    return np.array(results)
```

```python
import os
import multiprocessing

def run_one_simulation(values):
    with open(os.devnull, 'w') as f:
        with redirect_stdout(f):
            values_dict = {p.name: v for p, v in zip(variables, values)}
            values_dict.update(fixed_parameter_values)
            print(values_dict)
            try:
                result = component_temperature_gas_and_beam(**values_dict)
            except RuntimeError:
                result = np.nan
            print(result)
    return result

def run_many_simulations_parallel(data):
    """
    Running it in parallel like this makes it 4x faster on my 4-core macbook pro,
    but I have read rumours this may not work so well on Windows, as implemented here.
    Also, this method discards all logs, so to debug you may want the serial version above
    and then look at the 'simulations.log' file
    """
    num_cores = multiprocessing.cpu_count()
    print("Running on {} cores".format(num_cores))
    pool = multiprocessing.Pool(num_cores)
    results = pool.map(run_one_simulation, data)
    return np.array(results)
```

```python
def remove_nans(sample, output, number_of_samples):
    """
    Given a sample matrix, output, and number of samples,
    it returns the same three things, having removed any samples that gave
    numpy.nan as an output.
    """
    chunk_size = len(sample) / number_of_samples
    assert chunk_size == int(chunk_size), "Number of chunks doesn't correspond to this sample"
    chunk_size = int(chunk_size)
    assert len(output) == len(sample), "Output doesn't correspond to this sample"

    while np.any(np.isnan(output)):
        first_nan = np.nonzero(np.isnan(output))[0][0]
        chunk_index = int(first_nan/chunk_size)
        to_delete = range(chunk_size*chunk_index, chunk_size*(chunk_index+1))
        output = np.delete(output, to_delete)
        sample = np.delete(sample, to_delete, axis=0)
        number_of_samples = number_of_samples - 1
    return sample, output, number_of_samples
```

## Conclusions/recommendations so far

* should figure out pressure of nitrogen gas. Currently it's a super wide range.
* the size of the shelves used for surface area calculation in the gas convection heat transfer are also wide estimates (and strongly correlated with the uncertain gas pressure)
* probably these wide uncertainties are mostly driving the apparent importance of the gas phase heat transfer
* the dimensions of the beam (how far the heat has to diffuse, and how wide the cross-section it can diffuse down) are quite important and should be guessed better than I currently have.

# Next steps
I would recommend picking the bounds on the parameters much better, especially those that can be reasonably estimated (dimensions, pressure, etc.) then running many more simulations (increase the number of runs) and re-interpreting the results.

It would be nice to save the data and output matrices and present them using [parallel coordinates](https://en.wikipedia.org/wiki/Parallel_coordinates), like
http://mbostock.github.io/protovis/ex/cars.html or 
https://syntagmatic.github.io/parallel-coordinates/

Also, could include "fudge factor" variables to capture sensitivity to systematic uncertainty (eg. for using the wrong correlation or approximation, rather than uncertainty in a parameter).

Making the sensitivity analysis robust to at least a few NaN values would be wise.
